## Accepted Credit Card Offer
    We all recieve credit card offers at least every week, but how many of them are accepted every time? What does affect on our decision or how do we make our decision?<br>
    I am trying to get some idea to find the answer by making some plots, first I made two scatter plots as you see in the next, but I did not find any trend or relation, so I decided to use seaborn to see the factors and the trend, relation at one time. <br>
    There are two options for offer ; Accepted:Yes and Not accepted:No. As you will see there are two plot groups; Yes and No and also trend in Income Level and credit Rating in these groups.<br>
    In Yes group, the trend is going to High or Medium Income Level as # Credit Card Helds and # Bank Account are going to hingh. In No group, the trend is going to Low as # Credit Card Helds and # Bank Account are going to high.
    But another trend which is Credit Rating is more clear in both groups, in Yes group the trend is going to Medium Credit Rating and in No group is going to Low Rate.<br>
    The hypothesis: the relation between Acceping offer and Credit Rate is stronger than Income Level.


```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/home/eli/Downloads/creditcardmarketing-bbm.csv')
```

```python
df_dropped = df.dropna()
print(df.info())
```

```python

plt.scatter(df_dropped['Income Level'], df_dropped['# Credit Cards Held'])
plt.show()
```

```python
plt.scatter(df_dropped['Income Level'], df_dropped['Average Balance'], s=1)
plt.xlabel('Income Level')
plt.ylabel('average balance')
plt.show()
```

```python
df_Offer_Accepted = df_dropped[df_dropped['Offer Accepted'] =='Yes']
```

```python
sns.pairplot(df_Offer_Accepted, vars= ['Average Balance', '# Credit Cards Held', '# Bank Accounts Open', 'Household Size'], 
             hue='Income Level', plot_kws = {'alpha': 0.3, 's': 60})
```

```python
df_Offer_Notaccepted = df_dropped[df_dropped['Offer Accepted'] =='No']
sns.pairplot(df_Offer_Notaccepted, vars= ['Average Balance', '# Credit Cards Held', '# Bank Accounts Open', 'Household Size'], 
             hue='Income Level', plot_kws = {'alpha': 0.3, 's': 60})
```

```python
df_Offer_Notaccepted = df_dropped[df_dropped['Offer Accepted'] =='Yes']
sns.pairplot(df_Offer_Notaccepted, vars= ['Average Balance', '# Credit Cards Held', '# Bank Accounts Open', 'Household Size'], 
             hue='Credit Rating', plot_kws = {'alpha': 0.6, 's': 60})
```
